##Motivación
¿Por qué startups?
Primero, ¿qué hace una startup? Una startup está completamente enfocada en encontrar el product-market fit (AKA generar un producto o servicio de tal manera que tenga un lugar en el mercado y que haya clientes dispuestos a usarlo), lo cual se obtiene haciendo un intensivo análisis de mercado y concentrándose en construir una solución a un problema muy específico para un nicho de clientes en el comienzo. Dicho esto, una startup surge de uno o varios problemas y busca solucionarlos. Luego surge la pregunta: ¿En qué se diferencia una startup de una PyME o una empresa tradicional? Bueno, tal y como hace una empresa tradicional, una startup soluciona un problema, pero la startup carece de algo que tienen las empresas tradicionales al comienzo: un modelo de negocio bien conocido y probado. Las empresas tradicionales saben quiénes son sus clientes, cómo producir sus productos o servicios y cómo comercializarlos. Un fundador de startup con una idea loca e innovadora no sabe nada de eso.

Lo dicho anteriormente puede ser considerado como una desventaja de las startups con respecto a las empresas tradicionales, pero tiene una contraparte. Las startups remodelan el mundo, construyen el futuro. Tomemos como ejemplo a la startup más sexy de la actualidad, OpenAI. ¡Cambió el mundo! Es casi imposible cuantificarlo, pero cuántos trabajos nuevos generó, cómo aumentó la productividad de millones de clientes y también cuántos trabajos dejó y va a dejar obsoletos; eso entra en nuestra definición de cambiar el mundo. También pensemos en el caso de las startups fintech, como Mercado Libre (Mercado Pago, actualmente podria no considerarse una startup pero comenzó como una), Ualá, Nubank, y un largo etcétera. Estas empresas lograron bancarizar a los sectores más precarios de Latinoamérica, rezagados por la industria bancaria tradicional, dando crédito y facilitando las vidas de millones de personas.

Dicho todo esto, creemos que el camino para solucionar los problemas a los que se enfrenta la humanidad, prolongar la existencia de la raza humana y hacer del mundo un lugar mejor para las generaciones futuras es mediante la búsqueda constante de innovación para hacer cada día un mundo mejor. ¿Y qué mejor fuente de innovación que las startups?

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

X_train, X_test, y_train, y_test = train_test_split(data, y, test_size=0.25, random_state=42, stratify=y)

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print(f"Tamaño test entrenamiento: {X_train.shape}")
print(f"Tamaño test de prueba: {X_test.shape}")
```

Naive Bayes

Elegimos Naive Bayes como nuestro primer modelo debido a su simplicidad y rapidez. Aunque reconocemos sus limitaciones, como la asunción de independencia total entre variables y la normalidad de las mismas, su eficiencia y facilidad de implementación lo convierten en una excelente opción inicial para establecer una línea base en nuestro proyecto.

```python
from sklearn.naive_bayes import GaussianNB
def NB():
    model = GaussianNB()

    model.fit(X_train, y_train)


    y_pred = model.predict(X_test)

    print(classification_report(y_test, y_pred, target_names=target_names))

    labels = target_names
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    disp.plot()
NB()
```

Aqui vemos que Naive Bayes como modelo base tiene un accuracy del 31% mas o menos equivalente a clasificar al azar entre las tres clases, cabe resaltar que es la misma accuracy que elegir al azar la clase pero el clasificador en realidad elige la mayoria de las veces por clasificar a la startup como cerrada (closed). Usaremos modelos mas potentes para intentar mejorar esat clasificación.

Perceptron

Luego escogimos el modelo Perceptron. Este se trata de un modelo lineal que nos permite manejar mejor la clasificación cuando las variables no son totalmente independientes, como asume Naive Bayes.


```python
from sklearn.linear_model import Perceptron

def Percep():
    model = Perceptron()

    model.fit(X_train, y_train)


    y_pred = model.predict(X_test)

    print(classification_report(y_test, y_pred, target_names=target_names))

    labels = target_names
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    disp.plot()
Percep()
```

Si usamos como clasificador al perceptron, vemos que el accuracy mejora significativamente al obtenido usando naive bayes, nos da un 55%. A pesar de esto vemos que tiene un sesgo por clasificar el estado de las startups como operating. Al no dar una clasificación muy buena el perceptron, podemos concluir que las clases no son linealmente separables, pues si lo fueran el perceptron nos daria los hiperplanos de separación perfecta entre ellas. A pesar de esto nos da los hiperplanos optimos para las clases.

Logistic Regression con Cross Validation

Como tercer modelo, decidimos utilizar Regresión Logística con validación cruzada. Este enfoque con validación cruzada nos ayuda a evaluar y mejorar la robustez del modelo, garantizando que su rendimiento sea consistente y generalizable a nuevos datos

```python
from sklearn.linear_model import LogisticRegressionCV

def LRCV():
    model = LogisticRegressionCV(cv=5, penalty='l2',solver='liblinear', tol=1e-6, max_iter=int(1e6))

    model.fit(X_train, y_train)


    y_pred = model.predict(X_test)

    print(classification_report(y_test, y_pred, target_names=target_names))

    labels = target_names
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    disp.plot()
LRCV()
```

Podemos ver que usando logistic regression con cross validation se mejora el accuracy con respecto al perceptron, esat sube a un 61%. A pesar de esto se puede ver que tiene un sesgo por clasificar a las startups mayoritariamente como Aquired y Operating (las dos clases mayoritarias). Por lo que en modelos siguientes bsucaremos mejorar este punto

KNN\
Haremos un grid search para buscar el mejor numero de vecinos

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

neighbors = KNeighborsClassifier(n_neighbors=5)
parameters = {"n_neighbors":[1,5,10,15,20,25,30,35,40,45,50]}

clf = GridSearchCV(neighbors, parameters)

clf.fit(X_train, y_train)

scores = clf.cv_results_['mean_test_score']

plt.semilogx(parameters["n_neighbors"], scores)
plt.xlabel('n_neighbors')
plt.ylabel('accuracy')
plt.title('KNN Model Performance')
print(clf.best_params_)
print('accuracy =', clf.best_score_)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_classifier = KNeighborsClassifier(n_neighbors=clf.best_params_['n_neighbors'],)  # Puedes ajustar el número de vecinos según lo desees

# Entrenar el clasificador
knn_classifier.fit(X_train, y_train)

# Hacer predicciones sobre el conjunto de prueba
y_pred = knn_classifier.predict(X_test)

print(classification_report(y_test, y_pred, target_names=target_names))

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
disp.plot()
```

Bagging con Arboles de decisión como clasificadores debiles

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

# Ajustar el clasificador base
tree = DecisionTreeClassifier(criterion='entropy',
                              max_depth=8,
                              min_samples_split=12,
                              min_samples_leaf=6,
                              random_state=42)

# Ajustar el clasificador de bagging
bag = BaggingClassifier(estimator=tree,
                        n_estimators=700,
                        oob_score=True,
                        bootstrap=True,
                        max_samples=1.0,
                        n_jobs=-1,
                        random_state=42)

bag.fit(X_train, y_train)

print("Train Accuracy: %0.2f" % bag.score(X_train, y_train))
print("OOB Accuracy: %0.2f" % bag.oob_score_)
print("Test Accuracy: %0.2f" % bag.score(X_test, y_test))

y_pred = bag.predict(X_test)

print(classification_report(y_test, y_pred, target_names=target_names))

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
disp.plot()
```

En los clasificadores de mejor rendimiento, se puede observar que tienden a confundir startups cuya etiqueta real es "Closed" con "Acquired". Esto tiene una explicación: una startup que cierra es claramente una startup no exitosa, mientras que una startup que es adquirida puede parecer un éxito a primera vista, pero esto no es necesariamente así, ya que su éxito depende mucho del contexto. Un ejemplo de esto es Yahoo, que recibió una oferta de adquisición de Microsoft por 45 billones(americanos) de dólares en 2008, la cual fue rechazada. Años más tarde, en 2017, Yahoo fue adquirida por Verizon Communications por 4.48 billones(americanos) de dólares. Comparando esta cifra con la oferta de Microsoft, claramente se consideraría esta adquisición como un rotundo fracaso.

Por otro lado, hay startups que rechazaron adquisiciones, siguen operando y son consideradas un éxito. Este es el caso de Facebook, que con tan solo dos años de existencia, recibió una oferta de adquisición por 1 billon(americano) de dólares por parte de Yahoo, la cual fue rechazada. A día de hoy(10 de junio de 2024), Facebook (Meta) tiene un valor de mercado de 1.27 trillones(americanos) de dólares. Por lo tanto, la empresa rechazó una adquisición y resultó ser un éxito.

Con estos ejemplos no queremos decir que una empresa será exitosa si no es adquirida, ni que una adquisición es algo malo, solo queremos ilustrar que determinar si una startup es exitosa o no no es una tarea trivial, ya que depende mucho del contexto, de innumerables factores y, a su vez, del azar.

```python
df_para_binary['status'] = df_para_binary['status'].replace('acquired', 'closed') #Clasificare entre empresas que siguen operando y empresas que no.

#Balanceo de clases
df_operating = df_para_binary[df_para_binary['status'] == 'operating']
df_operating = df_operating.sample(n=5000, random_state=42)

df_not_operating = df_para_binary[df_para_binary['status'] != 'operating']

df_reduced_binary = pd.concat([df_operating, df_not_operating])

df_reduced_binary['status'].value_counts()

columnas_no_numericas = df_reduced_binary.select_dtypes(exclude=['number']).columns.tolist()

df_reduced_binary = pd.get_dummies(df_reduced_binary, prefix=None, prefix_sep='_', columns=["country_code_grouped","market"])

labels_bi = ['closed', 'operating']
target_bi = df_reduced_binary['status'].copy()
df_reduced_binary.drop(columns=['status'], inplace=True)

X_trainb, X_testb, y_trainb, y_testb = train_test_split(df_reduced_binary, target_bi, test_size=0.25, random_state=42, stratify=target_bi)


scaler = MinMaxScaler()
X_trainb = scaler.fit_transform(X_trainb)
X_testb = scaler.transform(X_testb)

# Ajustar el clasificador base
tree = DecisionTreeClassifier(criterion='entropy',
                              max_depth=None,
                              min_samples_split=12,
                              random_state=42)

# Ajustar el clasificador de bagging
bag = BaggingClassifier(estimator=tree,
                        n_estimators=300,
                        oob_score=True,
                        bootstrap=True,
                        max_samples=1.0,    # Usar el 100% de las muestras
                        n_jobs=-1,
                        random_state=42)

bag.fit(X_trainb, y_trainb)

print("Train Accuracy: %0.2f" % bag.score(X_trainb, y_trainb))
print("OOB Accuracy: %0.2f" % bag.oob_score_)
print("Test Accuracy: %0.2f" % bag.score(X_testb, y_testb))

y_pred = bag.predict(X_testb)

print(classification_report(y_testb, y_pred, target_names=labels_bi))

cm = confusion_matrix(y_testb, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels_bi)
disp.plot()
```
